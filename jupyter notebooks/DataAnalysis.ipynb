```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'file_path.csv' with the path to your CSV file
file_path0 = r'/home/ruediger/git/ALNS-and-RHACO-for-DVRP/urgency.csv'
file_path1 = r'/home/ruediger/git/ALNS-and-RHACO-for-DVRP/ALNS.csv'
file_path2 = r'/home/ruediger/git/ALNS-and-RHACO-for-DVRP/RHACO.csv'
```

```python
# Read the zeroth CSV file and add a source column
df0 = pd.read_csv(file_path0,sep=";")

# Read the first CSV file and add a source column
df1 = pd.read_csv(file_path1,sep=";")
df1['Heuristic'] = 'ALNS'  # Adding a column with the source name
df1 = pd.concat([df1, df0], axis=1)

# Read the second CSV file and add a source column
df2 = pd.read_csv(file_path2,sep=";")
df2['Heuristic'] = 'RHACO'  # Adding a column with the source name
df2 = pd.concat([df2, df0], axis=1)


# Concatenate the two DataFrames into one
df = pd.concat([df1, df2], axis=0, ignore_index=True)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
print(df.columns)
```

```python
heuristics = df['Heuristic'].unique()
colors = ['tab:blue', 'tab:red']  # Adjust the color list as needed

# Create a color dictionary
color_dict = dict(zip(heuristics, colors))

plt.figure(figsize=(8, 5))
for heuristic, color in color_dict.items():
    # Filter the data by category
    heuristic_data = df[df['Heuristic'] == heuristic]
    plt.scatter(heuristic_data['dynamic'], heuristic_data['Funktionswert'], color=color, label=heuristic)

# Adding plot title and labels
#plt.title('Scatter Plot Colored by Category')
plt.xlabel('Dynamik')
plt.ylabel('Funktionswert')
plt.legend(title='Heuristik')
plt.savefig('Funktionswert zu Dynamik.png', format='png', dpi=300)
plt.show()
```

```python
grouped_RHACO = df[df['Heuristic']=="RHACO"].groupby('dod')['Funktionswert'].agg([pd.Series.mean, pd.Series.std])
grouped_RHACO.columns = ['Mean', 'Std']  # Renaming the columns for clarity
grouped_ALNS = df[df['Heuristic']=="ALNS"].groupby('dod')['Funktionswert'].agg([pd.Series.mean, pd.Series.std])
grouped_ALNS.columns = ['Mean', 'Std']  # Renaming the columns for clarity


plt.figure(figsize=(8, 5))
plt.errorbar(grouped_ALNS.index, grouped_ALNS['Mean'], fmt=':s', yerr=grouped_ALNS['Std'], capsize=5, color='tab:blue', markersize=5, label='ALNS')
plt.errorbar(grouped_RHACO.index, grouped_RHACO['Mean'], fmt=':s', yerr=grouped_RHACO['Std'], capsize=5, color='tab:red', markersize=5, label='RHACO')

  



plt.legend(title='Heuristik')
#plt.title('Mean Values by Category with Standard Deviation Error Bars')
plt.xlabel('DoD')
plt.ylabel('Funktionswert')

plt.savefig('DOD error bars.png', format='png', dpi=300)
plt.show()
```
